### Extract and save selected chain

```python
from Bio.PDB import PDBIO, Select

class ChainSelect(Select):
    def __init__(self, target_chain):
        self.target_chain = target_chain

    def accept_model(self, model):
        return True  # Accept the first (and usually only) model

    def accept_chain(self, chain):
        return chain.id == self.target_chain

    def accept_residue(self, residue):
        return True  # Accept all residues in the selected chain

    def accept_atom(self, atom):
        return True  # Accept all atoms

def save_chain_as_pdb(structure, chain_id, output_filename="selected_chain.pdb"):
    io = PDBIO()
    io.set_structure(structure)
    io.save(output_filename, select=ChainSelect(chain_id))
    print(f"‚úÖ Saved chain '{chain_id}' to '{output_filename}'")


save_chain_as_pdb(structure, selected_chain_id)
```

### Remove heteroatoms and water

```python
from Bio.PDB import PDBParser, PDBIO, Select

# Define a selector that removes heteroatoms, water, and alternate locations
class CleanProteinSubunit(Select):
    def accept_residue(self, residue):
        hetfield, resseq, icode = residue.get_id()
        return hetfield == " " and residue.get_resname() != "HOH"

    def accept_atom(self, atom):
        return not atom.is_disordered()  # Removes altLoc atoms

def clean_protein_subunit(input_pdb, output_pdb="protein.pdb"):
    parser = PDBParser(QUIET=True)
    structure = parser.get_structure("protein", input_pdb)

    io = PDBIO()
    io.set_structure(structure)
    io.save(output_pdb, select=CleanProteinSubunit())

    print(f"‚úÖ Cleaned subunit saved as '{output_pdb}'")


clean_protein_subunit("selected_chain.pdb")
```

# Preparation for Docking and Docking

## Automate Docking with AutoDock Vina

* Set the docking box coordinates (center & size)

* Create the config file for Vina

* Run Vina from Python (using subprocess)

### Calculate Ligand Centre

```python
from rdkit import Chem
from rdkit.Chem import AllChem

def calculate_ligand_center(pdb_file):
    mol = Chem.MolFromPDBFile(pdb_file, removeHs=False)
    if mol is None:
        raise ValueError("Could not parse ligand file.")

    conf = mol.GetConformer()
    num_atoms = mol.GetNumAtoms()

    x, y, z = 0.0, 0.0, 0.0
    for i in range(num_atoms):
        pos = conf.GetAtomPosition(i)
        x += pos.x
        y += pos.y
        z += pos.z

    return x / num_atoms, y / num_atoms, z / num_atoms

Center_x, Center_y, Center_z = calculate_ligand_center("ligand.pdb")
print(f"Center of ligand: ({Center_x}, {Center_y}, {Center_z})")
```

### Create the Docking Box (config.txt)

```python
config_text = f"""
receptor = protein.pdbqt
ligand = ligand.pdbqt

center_x = {Center_x}
center_y = {Center_y}
center_z = {Center_z}

size_x = {Size_x}
size_y = {Size_y}
size_z = {Size_z}

out = output.pdbqt
log = log.txt
"""

with open("config.txt", "w") as f:
    f.write(config_text.strip())
```

# Post-Dock Analysis

## Extract Binding Affinities from output.pdbqt

```python
# Extract binding affinities from output.pdbqt
affinities = []

with open("output.pdbqt") as f:
    for line in f:
        if line.startswith("REMARK VINA RESULT:"):
            parts = line.strip().split()
            if len(parts) >= 4:
                affinity = float(parts[3])  # e.g., -7.4
                affinities.append(affinity)

if affinities:
    print("Binding affinities (kcal/mol):")
    for i, score in enumerate(affinities):
        print(f"Pose {i+1}: {score}")
else:
    print("No Vina results found.")
```

## Convert output.pdbqt to output.pdb using Open Babel

```python
import subprocess

# Convert output.pdbqt to PDB (for visualization)
try:
    result = subprocess.run(
        ["timeout", "10", "obabel", "-ipdbqt", "output.pdbqt", "-O", "output.pdb"],
        check=True,
        capture_output=True,
        text=True
    )
    print("Conversion successful. output.pdb is ready.")
except subprocess.CalledProcessError as e:
    print("Conversion failed. Error:")
    print(e.stderr)
```

## Visualise results using Py3Dmol

```python
import py3Dmol

# Load and visualize docking result (output.pdb)
with open('output.pdb', 'r') as f:
    pdb_data = f.read()

view = py3Dmol.view(width=800, height=600)
view.addModel(pdb_data, "pdb")
view.setStyle({'stick': {}})
view.setBackgroundColor('white')
view.zoomTo()
view.show()
```
